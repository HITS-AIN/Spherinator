```python
from spherinator.data import DataModule, Column

loader = DataModule(
    path="HITS-AIN/Illustris_TNG_SKIRT_SDSS",
    columns=["image", "subhalo_id", "simulation", "snapshot"],
    validation_size=0.0,
    batch_size=10,
    shuffle=False,
)
loader.setup()
```

```python
import matplotlib.pyplot as plt

batch = next(iter(loader.train_dataloader()))
print(batch.keys())

images = batch["image"][:10]
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i].numpy().transpose(1, 2, 0))
    ax.axis("off")
plt.tight_layout()
plt.show()
```
