```python
import pandas as pd
import numpy as np
import numpy as np
```

```python

file = "/home/doserbd/data/gaia/xp_sampled_mean_spectrum/csv/XpSampledMeanSpectrum_000000-003111.csv"
gaia_sampled = pd.read_csv(file, comment='#')
gaia_sampled
```

```python

list_of_arrays = [
    "flux",
    "flux_error",
]

for array in list_of_arrays:
    gaia_sampled[array] = gaia_sampled[array].apply(
        lambda x: np.fromstring(x[1:-1], dtype=np.float32, sep=",")
    )
```

```python
file = "/home/doserbd/data/gaia/xp_continuous_mean_spectrum/csv/XpContinuousMeanSpectrum_000000-003111.csv.gz"
gaia_continuous = pd.read_csv(file, comment='#')
gaia_continuous
```

```python
list_of_arrays = [
    "bp_coefficients",
    "bp_coefficient_errors",
    "bp_coefficient_correlations",
    "rp_coefficients",
    "rp_coefficient_errors",
    "rp_coefficient_correlations",
]

for array in list_of_arrays:
    gaia_continuous[array] = gaia_continuous[array].apply(
        lambda x: np.fromstring(x[1:-1], dtype=np.float32, sep=",")
    )
```

```python
import matplotlib.pyplot as plt

plt.plot(range(gaia_sampled["flux"][0].shape[0]), gaia_sampled["flux"][0])
plt.show()
```

All mean spectra are sampled to the same set of absolute wavelength positions, viz. 343 values from
336 to 1020 nm with a step of 2 nm.

Source: https://gea.esac.esa.int/archive/documentation/GDR3/Gaia_archive/chap_datamodel/sec_dm_spectroscopic_tables/ssec_dm_xp_sampled_mean_spectrum.html

```python
from gaiaxpy import calibrate

calibrated_spectrum, sampling = calibrate(
    gaia_continuous, sampling=np.arange(336, 1021, 2), save_file=False
)
```

```python
plt.plot(range(calibrated_spectrum["flux"][1].shape[0]), calibrated_spectrum["flux"][1])
plt.show()
```

```python
df = gaia_continuous.loc[1:1]
df["bp_coefficients"]
```

```python
max_bp = np.max(df["bp_coefficients"].to_numpy()[0])
max_rp = np.max(df["rp_coefficients"].to_numpy()[0])
min_bp = np.min(df["bp_coefficients"].to_numpy()[0])
min_rp = np.min(df["rp_coefficients"].to_numpy()[0])
max = np.max([max_bp, max_rp])
min = np.min([min_bp, min_rp])
min, max
```

```python
abs_max = np.max([np.abs(min), np.abs(max)])
abs_max
```

```python
df["bp_coefficients"] = df["bp_coefficients"].apply(lambda x: x / abs_max)
df["rp_coefficients"] = df["rp_coefficients"].apply(lambda x: x / abs_max)
```

```python

calibrated_spectrum, sampling = calibrate(
    df, sampling=np.arange(336, 1021, 2), save_file=False
)
```

```python
plt.plot(range(calibrated_spectrum["flux"][0].shape[0]), calibrated_spectrum["flux"][0])
plt.show()
```
